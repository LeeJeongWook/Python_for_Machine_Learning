###6\. lambda, map, apply

**Lambda 함수**
- 한 줄로 함수를 표현하는 익명 함수 기법
- Lisp 언어에서 시작된 기법으로 오늘날 현대언어에 많이 사용

```python
f = lambda x,y: x + y
f(1,4)
```

```python
def f(x , y):
    return x + y
f(1,4)
```

```python
f = lambda x: x ** 2
f(3)
```

```python
f = lambda x: x+5
f(3)
```

```python
(lambda x: x +1)(5)     # 이름을 할당하지 않는 lambda 함수
```

```python
ex = [1,2,3,4,5]
f = lambda x: x ** 2
list(map(f, ex))
```

```python
f = lambda x, y: x + y
list(map(f, ex, ex))
```

```python
list(map(lambda x: x+5, ex)) 
#python 3에는 list를 꼭 붙여줘야함
```

**apply for dataframe**
- map과 달리, series 전체(column)에 해당 함수를 적용
- 입력값이 series 데이터로 입력받아 handling 가능

```python
f = lambda x : x.max() - x.min()
df_info.apply(f)
```

- 내장 연산 함수를 사용할 때도 똑같은 효과를 거둘 수 있음
- mean, std 등 사용가능

```python
df_info.apply(sum)
```

```python
df_info.sum()
```

```python
def f(x):
    return Series([x.min(), x.max(), x.mean()], 
                    index=["min", "max", "mean"])
df_info.apply(f)
```

**unique**
- series data의 유일한 값을 list를 반환함

```python
df.race.unique()    # 유일한 인종의 값 list
```

```python
np.array(dict(enumerate(df["race"].unique())))  #dict type으로 index 
```

```python
value = list(map(int, np.array(list(enumerate(df["race"].unique())))[:,0].tolist()))
key = np.array(list(enumerate(df["race"].unique())), dtype=str)[:,1].tolist()
print(value)
print(key)      # label index 값과 label 값 각각 추출
```

```python
df["race"].replace(to_replace=key, value=value, inplace=True)
```

```python
value = list(map(int, np.array(list(enumerate(df["sex"].unique())))[:,0].tolist()))
key = np.array(list(enumerate(df["sex"].unique())), dtype=str)[:, 1].tolist()

value, key
```

```python
df["sex"].replace(to_replace=key, value=value, inplace=True)
df.head(5)
```
